```python
import re

def start():
  global choice
  print('Welcome to the process')
  choice = input('Do you want to Register/Login? ')
  #register = ['Register', 'register', 'REGISTER']
  #login = ['login', 'LOGIN', 'Login']
  if (choice != 'Register' and choice != 'Login'):
    print('Please enter either Register or Login')
    start()

def access(choice):
  if choice == 'Login':
    name = input('Enter your username: ')
    password = input('Enter your password: ')
    login(name, password)
  else:
    print('Enter your mail id and password to register')
    name = input('Enter your e-mail id: ')
    regex = '^[a-z]+[\._]?[a-z0-9]+[@]\w+[.]\w{2,3}$'
    if (re.search(regex, name)):
      register_password(name)
    else:
      print('Your email id is wrong, please restart the registration')
      access(choice)

## Registration functions

def register_password(name):
  password = input('Enter new password: ')
  reg = "^(?=.*[a-z])(?=.*[A-Z])(?=.*\d)(?=.*[@$!%*#?&])[A-Za-z\d@$!#%*?&]{6,20}$"
  if (re.search(reg, password)) or (5<=len(password)>=16):
    register(name, password)
  else:
    print("{}\n{}".format(
    "You have entered wrong password, please enter as below standards",
    "Must be of length between 5-16, one capital letter, one small letter and one special character",
    ))
    register_password(name)

def register(name, password):
  file = open('user_details.txt', 'a')
  file.write("\n"+name+","+password)
  file.close()
  print('Registration completed')

    
## Login functions

def login(name, password):
  success = False
  try:
   file = open('user_details.txt', 'r')
   for i in file:
     b = i.split(",")
     if (b[0] == name and b[1] == password):
       success = True
       break
   file.close()
   if (success):
     print('Login successful')
   else:
     print('Wrong username or password')
     login_choice(name, password)
  except FileNotFoundError:
    print('Database not found, please register from first')
    start()
    access(choice)
    
  
def login_choice(name, password):
  print("{}\n{}\n{}".format(
      "Choose 1 or 2",
      "1. Register",
      "2. Forgot Password"
  ))
  option = input('Choose 1 or 2: ')
  if option == str(1):
    print('You choose Register')
    start()
    access(choice)
  elif option == str(2):
    print('Enter your registered username (email id): ')
    registered_name = input('Enter your username: ')
    file = open('user_details.txt', 'r')
    for i in file:
      b = i.split(",")
      if (b[0] == registered_name):
        print('Your password is: ', b[1])
        break
    else:
      print('You have entered wrong login username')
      start()
      access(choice)
  else:
    print('Please choose either 1 or 2 ')
    login_choice(name, password)

def function_call():
    start()
    access(choice)

function_call()
```
